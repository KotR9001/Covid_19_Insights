# Clean Data

```python
#Select Features for Counties
counties_features = counties[['FIPS', 'State', 'Area_Name',
                              'Rural-urban_Continuum Code_2013', 'Urban_Influence_Code_2013', 'Economic_typology_2015',
                             'POP_ESTIMATE_2018', 'Percent of adults with less than a high school diploma 2014-18', 
                             'Percent of adults with a high school diploma only 2014-18', "Percent of adults completing some college or associate's degree 2014-18",
                             "Percent of adults with a bachelor's degree or higher 2014-18", 'Unemployment_rate_2018',
                             'Median_Household_Income_2018', 'Med_HH_Income_Percent_of_State_Total_2018',
                             'Mar Temp Min / F', 'Density per square mile of land area - Population',
                             'Density per square mile of land area - Housing units', 'Active Physicians per 100000 Population 2018 (AAMC)',
                             'Total Active Patient Care Physicians per 100000 Population 2018 (AAMC)',
                             'Active Primary Care Physicians per 100000 Population 2018 (AAMC)',
                             'Active Patient Care Primary Care Physicians per 100000 Population 2018 (AAMC)',
                             'Active General Surgeons per 100000 Population 2018 (AAMC)',
                             'Active Patient Care General Surgeons per 100000 Population 2018 (AAMC)',
                             'MD and DO Student Enrollment per 100000 Population AY 2018-2019 (AAMC)',
                             'Student Enrollment at Public MD and DO Schools per 100000 Population AY 2018-2019 (AAMC)',
                             'Total Residents/Fellows in ACGME Programs per 100000 Population as of December 31 2018 (AAMC)',
                             'Total Residents/Fellows in Primary Care ACGME Programs per 100000 Population as of Dec. 31 2018 (AAMC)',
                             'ICU Beds', 'crime_rate_per_100000']]
counties_features = counties_features.dropna()
counties_features
```

```python
#Create Grocery Visits Pivot Table
grocery_visits_pivot = pd.melt(grocery_visits, id_vars='FIPS', var_name="date", value_name='Grocery_Visits')
grocery_visits_pivot_cleaned = grocery_visits_pivot.reset_index(drop=True)
grocery_visits_pivot_cleaned['date'] = pd.to_datetime(grocery_visits_pivot_cleaned['date'])
grocery_visits_pivot_cleaned
```

```python
#Create Healthcare Visits Pivot Table
healthcare_visits_pivot = pd.melt(healthcare_visits, id_vars='FIPS', var_name="date", value_name='Healthcare_Visits')
healthcare_visits_pivot_cleaned = healthcare_visits_pivot.reset_index(drop=True)
healthcare_visits_pivot_cleaned['date'] = pd.to_datetime(healthcare_visits_pivot['date'])
healthcare_visits_pivot_cleaned
```

```python
#Create Hospital Visits Pivot Table
hospital_visits_pivot = pd.melt(hospital_visits, id_vars='FIPS', var_name="date", value_name='Hospital_Visits')
hospital_visits_pivot_cleaned = hospital_visits_pivot.reset_index(drop=True)
hospital_visits_pivot_cleaned['date'] = pd.to_datetime(hospital_visits_pivot_cleaned['date'])
hospital_visits_pivot_cleaned
```

```python
#Create Interventions Pivot Table
#Found Method to Convert Julian to Gregorian Dates from https://oneau.wordpress.com/2011/08/30/jdcal/
#and https://stackoverflow.com/questions/48587595/convert-julian-dates-to-normal-dates-in-a-dataframe
```

```python
#Create POI Visits Pivot Table
poi_visits_pivot = pd.melt(poi_visits, id_vars='FIPS', var_name="date", value_name='POI_Visits')
poi_visits_pivot_cleaned = poi_visits_pivot.reset_index(drop=True)
poi_visits_pivot_cleaned['date'] = pd.to_datetime(poi_visits_pivot_cleaned['date'])
poi_visits_pivot_cleaned
```

```python
#Create Infections Pivot Table
infections_pivot = pd.melt(infections_timeseries, id_vars=['FIPS', 'Combined_Key'], var_name="date", value_name='Number of Infections')
infections_pivot_filtered = infections_pivot.loc[(infections_pivot['date']>='3/0/2020') & (infections_pivot['date']<='3/30/2020')]
infections_pivot_filtered['date'] = pd.to_datetime(infections_pivot_filtered['date'])
infections_pivot_cleaned = infections_pivot_filtered.sort_values(['FIPS', 'date'], ascending=True).reset_index(drop=True)
infections_pivot_cleaned
```

```python
#Create Deaths Table
deaths_pivot = pd.melt(deaths_timeseries, id_vars=['FIPS', 'Combined_Key'], var_name="date", value_name='Number of Deaths')
deaths_pivot_filtered = deaths_pivot.loc[(infections_pivot['date']>='3/0/2020') & (infections_pivot['date']<='3/30/2020')]
deaths_pivot_filtered['date'] = pd.to_datetime(deaths_pivot_filtered['date'])
deaths_pivot_cleaned = deaths_pivot_filtered.sort_values(['FIPS', 'date'], ascending=True).reset_index(drop=True)
deaths_pivot_cleaned
```

# Join Tables

```python
#Join Outputs Tables
infections_deaths_pivot = infections_pivot_cleaned.join(deaths_pivot_cleaned['Number of Deaths'], how='inner', on='FIPS')
infections_deaths_pivot_sorted = infections_deaths_pivot.sort_values(['FIPS', 'date'], ascending=True).reset_index(drop=True)
infections_deaths_pivot_sorted
```

```python
#Join Feature Tables
#Found Method to Merge & Fill Missing Values at https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.merge_ordered.html
counties_grocery_pivots = pd.merge_ordered(counties_features, grocery_visits_pivot_cleaned, how='outer', on='FIPS', fill_method='ffill')
counties_grocery_pivots_sorted = counties_grocery_pivots.sort_values(['FIPS', 'date'], ascending=True).reset_index(drop=True)
counties_grocery_pivots_columns = counties_grocery_pivots_sorted.columns
counties_grocery_pivots_sorted = counties_grocery_pivots[['FIPS', 'State', 'Area_Name', 'date', 'Rural-urban_Continuum Code_2013',
       'Urban_Influence_Code_2013', 'Economic_typology_2015',
       'POP_ESTIMATE_2018',
       'Percent of adults with less than a high school diploma 2014-18',
       'Percent of adults with a high school diploma only 2014-18',
       "Percent of adults completing some college or associate's degree 2014-18",
       "Percent of adults with a bachelor's degree or higher 2014-18",
       'Unemployment_rate_2018', 'Median_Household_Income_2018',
       'Med_HH_Income_Percent_of_State_Total_2018', 'Mar Temp Min / F',
       'Density per square mile of land area - Population',
       'Density per square mile of land area - Housing units',
       'Active Physicians per 100000 Population 2018 (AAMC)',
       'Total Active Patient Care Physicians per 100000 Population 2018 (AAMC)',
       'Active Primary Care Physicians per 100000 Population 2018 (AAMC)',
       'Active Patient Care Primary Care Physicians per 100000 Population 2018 (AAMC)',
       'Active General Surgeons per 100000 Population 2018 (AAMC)',
       'Active Patient Care General Surgeons per 100000 Population 2018 (AAMC)',
       'MD and DO Student Enrollment per 100000 Population AY 2018-2019 (AAMC)',
       'Student Enrollment at Public MD and DO Schools per 100000 Population AY 2018-2019 (AAMC)',
       'Total Residents/Fellows in ACGME Programs per 100000 Population as of December 31 2018 (AAMC)',
       'Total Residents/Fellows in Primary Care ACGME Programs per 100000 Population as of Dec. 31 2018 (AAMC)',
       'ICU Beds', 'crime_rate_per_100000', 'Grocery_Visits']]
counties_grocery_pivots_sorted
#counties_grocery_pivots_sorted.loc[(counties_grocery_pivots_sorted['date']=='2020-03-02')].dropna()
#len(counties_grocery_pivots_sorted['FIPS'].unique())
```

```python
#Join with Heathcare Visits
co_gr_hc_pivot = counties_grocery_pivots_sorted.merge(healthcare_visits_pivot_cleaned, how='inner', on=['FIPS', 'date'])
co_gr_hc_pivot_sorted = co_gr_hc_pivot.sort_values(['FIPS', 'date'], ascending=True).reset_index(drop=True)
co_gr_hc_pivot_sorted
```

```python
#Join with Hospital Visits
co_gr_hc_ho_pivot = co_gr_hc_pivot_sorted.merge(hospital_visits_pivot_cleaned, how='inner', on=['FIPS', 'date'])
co_gr_hc_ho_pivot_sorted = co_gr_hc_ho_pivot.sort_values(['FIPS', 'date'], ascending=True).reset_index(drop=True)
co_gr_hc_ho_pivot_sorted
```

```python
#Join with POI Visits
co_gr_hc_ho_poi_pivot = co_gr_hc_ho_pivot_sorted.merge(poi_visits_pivot_cleaned, how='inner', on=['FIPS', 'date'])
co_gr_hc_ho_poi_pivot_sorted = co_gr_hc_ho_poi_pivot.sort_values(['FIPS', 'date'], ascending=True).reset_index(drop=True)
co_gr_hc_ho_poi_pivot_sorted
```

```python
#Join with Interventions
co_gr_hc_ho_poi_in_pivot = co_gr_hc_ho_poi_pivot_sorted.merge(interventions, how='inner', on='FIPS')
co_gr_hc_ho_poi_in_pivot_sorted = co_gr_hc_ho_poi_in_pivot.sort_values(['FIPS', 'date'], ascending=True).reset_index(drop=True)
co_gr_hc_ho_poi_in_pivot_sorted
```

```python
#Join Features & Outputs Tables
master_df = co_gr_hc_ho_poi_in_pivot_sorted.merge(infections_deaths_pivot_sorted[['FIPS', 'date', 'Number of Infections',
                                                                             'Number of Deaths']], how='inner', on=['FIPS', 'date'])
master_df_cleaned = master_df.dropna(subset=['State', 'Area_Name'], how='any')
master_df_cleaned.dtypes
```

```python
#Import Machine Learning Module(s)
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import category_encoders as ce
```

```python
#Convert String Values in Categorical Columns to Labels
#Found Method at https://contrib.scikit-learn.org/category_encoders/
#Create the Model(s)
glmm = ce.glmm.GLMMEncoder(verbose=0, cols=['State', 'Area_Name', 'STATE', 'AREA_NAME'], return_df=True)
#Train the Models
glmm_infections_train = glmm.fit(x_train, y_train['Number of Infections'])
glmm_deaths_train = glmm.fit(x_train, y_train['Number of Deaths'])
glmm_infections_test = glmm.fit(x_test, y_test['Number of Infections'])
glmm_deaths_test = glmm.fit(x_test, y_test['Number of Deaths'])
```

```python
#Assign Variables for Encoded Data
x_train_enc = glmm_train_tf
x_test_enc = glmm_test_tf
```

```python
#Convert Date Column to Numerical Type
#Did this to Enable Linear Regression Model to be Trained
#Found Method at https://stackoverflow.com/questions/60049059/python-linear-regression-typeerror-invalid-type-promotion
x_train_enc['date'] = x_train_enc['date'].map(dt.datetime.toordinal)
x_test_enc['date'] = x_test_enc['date'].map(dt.datetime.toordinal)
```

```python
# Import StandardScaler to prepare data for Fitting to Avoid Biasing the Model to the Testing Data
from sklearn.preprocessing import StandardScaler

#Reshape the Data to Obtain a 2-D Array for the y_scaler
num_infections_train = y_train['Number of Infections'].values.reshape(-1, 1)
num_deaths_train = y_train['Number of Deaths'].values.reshape(-1, 1)
num_infections_test = y_test['Number of Infections'].values.reshape(-1, 1)
num_deaths_test = y_test['Number of Deaths'].values.reshape(-1, 1)

#Scale to Fit a Gaussian Distribution to the Data
x_scaler = StandardScaler().fit(x_train_enc)
y_scaler = StandardScaler().fit(num_infections_train)

#Transform the Data
x_train_scaled = pd.DataFrame(x_scaler.transform(x_train_enc), columns=x_train_enc.columns)
x_test_scaled = pd.DataFrame(x_scaler.transform(x_test_enc), columns=x_test_enc.columns)
y_train_scaled_infections = pd.DataFrame(y_scaler.transform(num_infections_train), columns=['Number of Infections'])
y_train_scaled_deaths = pd.DataFrame(y_scaler.transform(num_deaths_train), columns=['Number of Deaths'])
y_test_scaled_infections = pd.DataFrame(y_scaler.transform(num_infections_test), columns=['Number of Infections'])
y_test_scaled_deaths = pd.DataFrame(y_scaler.transform(num_deaths_test), columns=['Number of Deaths'])
```

```python
#Import Imputer to Fill in Missing Values
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
#Impute Values with Imputer Strategy
imputer = IterativeImputer(missing_values=np.nan, n_nearest_features=5, initial_strategy='mean')
x_train_enc_imp_tf = imputer.fit(x_train_scaled).transform(x_train_scaled)
x_train_enc_imp_tf
```

```python
x_test_enc_imp_tf = imputer.fit(x_test_scaled).transform(x_test_scaled)
x_test_enc_imp_tf
```

```python
#Train the Model to the Training Data
reg_infections = linear_reg.fit(x_train_enc_imp_tf, y_train_scaled_infections)
reg_deaths = linear_reg.fit(x_train_enc_imp_tf, y_train_scaled_deaths)

#Score the Models
score_infections = linear_reg.score(x_train_enc_imp_tf, y_train_scaled_infections)
score_deaths = linear_reg.score(x_train_enc_imp_tf, y_train_scaled_deaths)

print(f"The R^2 score for the infection predictions: {score_infections}")
print(f"The R^2 score for the death predictions: {score_deaths}")
```

```python
#Obtain Model Predictions
prediction_infections = linear_reg.predict(x_test_enc_imp_tf)[:, 0]
prediction_deaths = linear_reg.predict(x_test_enc_imp_tf)[:, 0]
len(prediction_infections)
```
