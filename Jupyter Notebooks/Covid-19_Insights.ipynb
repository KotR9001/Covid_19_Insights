```python
#Create Grocery Visits Pivot Table
grocery_visits_pivot = pd.melt(grocery_visits, id_vars='FIPS', var_name="date", value_name='Grocery_Visits')
grocery_visits_pivot_cleaned = grocery_visits_pivot.reset_index(drop=True)
grocery_visits_pivot_cleaned
```

```python
#Create Healthcare Visits Pivot Table
healthcare_visits_pivot = pd.melt(healthcare_visits, id_vars='FIPS', var_name="date", value_name='Healthcare_Visits')
healthcare_visits_pivot_cleaned = healthcare_visits_pivot.reset_index(drop=True)
healthcare_visits_pivot_cleaned
```

```python
#Create Hospital Visits Pivot Table
hospital_visits_pivot = pd.melt(hospital_visits, id_vars='FIPS', var_name="date", value_name='Hospital_Visits')
hospital_visits_pivot_cleaned = hospital_visits_pivot.reset_index(drop=True)
hospital_visits_pivot_cleaned
```

```python
#Create Interventions Pivot Table
#Found Method to Convert Julian to Gregorian Dates from https://oneau.wordpress.com/2011/08/30/jdcal/
#and https://stackoverflow.com/questions/48587595/convert-julian-dates-to-normal-dates-in-a-dataframe
```

```python
#Create POI Visits Pivot Table
poi_visits_pivot = pd.melt(poi_visits, id_vars='FIPS', var_name="date", value_name='POI_Visits')
poi_visits_pivot_cleaned = poi_visits_pivot.reset_index(drop=True)
poi_visits_pivot_cleaned
```

```python
#Create Infections Pivot Table
infections_pivot = pd.melt(infections_timeseries, id_vars=['FIPS', 'Combined_Key'], var_name="date", value_name='Number of Infections')
infections_pivot_filtered = infections_pivot.loc[(infections_pivot['date']>='3/0/2020') & (infections_pivot['date']<='3/30/2020')]
infections_pivot_cleaned = infections_pivot_filtered.reset_index(drop=True)
infections_pivot_cleaned
```

```python
#Create Deaths Table
deaths_pivot = pd.melt(deaths_timeseries, id_vars=['FIPS', 'Combined_Key'], var_name="date", value_name='Number of Deaths')
deaths_pivot_filtered = deaths_pivot.loc[(infections_pivot['date']>='3/0/2020') & (infections_pivot['date']<='3/30/2020')]
deaths_pivot_cleaned = deaths_pivot_filtered.reset_index(drop=True)
deaths_pivot_cleaned
```
